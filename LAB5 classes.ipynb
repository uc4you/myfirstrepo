```python
class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __add__(self,other):
        return(Vector(self.x + other.x, self.y + other.y))

    def __str__(self):
        return ("Vector: x: " + str(self.x) + " y: " + str(self.y))

    
v = Vector(1,2)
v1 = Vector(2,3)
v2 = v + v1

print (v2)
```
